# Data Exploration




```
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/NikoletaEm/datasps/refs/heads/main/NYPD_Shooting_Incident_Data__Historic_.csv"
viol = pd.read_csv(url)

# Display first few rows
viol.head()

```

**Check Data Structure**

```
viol.info()
```

**Summary Statistics**


```
viol.describe()
```

**Check for Missing Values**

```
viol.isnull().sum()
```

Columns like LOC_OF_OCCUR_DESC and LOC_CLASSFCTN_DESC have 25,596 missing values—which is  a significant portion of the dataset. Dropping them is a reasonable choice!

```
viol.drop(["LOC_OF_OCCUR_DESC", "LOC_CLASSFCTN_DESC"], axis=1, inplace=True)
```

We will also handle the missing values in the other columns by replacing them with Uknown

```
viol.fillna({"PERP_AGE_GROUP": "Unknown", "PERP_SEX": "Unknown", "PERP_RACE": "Unknown"}, inplace=True)
viol["LOCATION_DESC"].fillna("Unknown", inplace=True)
viol.dropna(subset=["Latitude", "Longitude"], inplace=True)
viol.isnull().sum()

```

**Convert Date and Time Columns**

```
viol["OCCUR_DATE"] = pd.to_datetime(viol["OCCUR_DATE"])
viol["OCCUR_TIME"] = pd.to_datetime(viol["OCCUR_TIME"], format="%H:%M:%S").dt.time
```

## **Plot 1: Heatmap**

My first plot will be a NYC heatmap visualizing the total number of shootings in each borough.

```
import pandas as pd
import geopandas as gpd
import folium
from branca.element import Template, MacroElement


borough_shootings = viol["BORO"].value_counts().reset_index()
borough_shootings.columns = ["borough", "shooting_count"]


nyc_geo = gpd.read_file("https://raw.githubusercontent.com/dwillis/nyc-maps/master/boroughs.geojson")


nyc_geo["borough"] = nyc_geo["BoroName"].str.upper()


nyc_geo = nyc_geo.merge(borough_shootings, on="borough", how="left").fillna(0)


m = folium.Map(location=[40.7128, -74.0060], zoom_start=11)


choropleth = folium.Choropleth(
    geo_data=nyc_geo,
    name="Shootings Heatmap",
    data=nyc_geo,
    columns=["borough", "shooting_count"],
    key_on="feature.properties.borough",
    fill_color="viridis",  # Colorblind-friendly palette
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Total Shootings per Borough",
).add_to(m)

# Custom CSS for making the legend bigger
legend_html = """
<style>
    .leaflet-control .legend {
        font-size: 16px !important;  /* Increase font size */
        padding: 10px 20px !important;  /* Add padding */
    }
</style>
"""


macro = MacroElement()
macro._template = Template(legend_html)
m.get_root().add_child(macro)


m
```

The choropleth map visualizes the total number of shootings per borough, with colors representing the intensity of shootings. The fill_color="viridis" property was used for a colorblind-friendly palette. The legend's size and padding were customized using CSS (font-size: 16px, padding: 10px 20px) to enhance readability.



## **Plot 2: Scatter plot**

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


np.random.seed(42)
num_shootings = 500


dates = pd.date_range(start="2023-01-01", periods=num_shootings, freq="D").to_list()
boroughs = np.random.choice(["BROOKLYN", "BRONX", "QUEENS", "MANHATTAN", "STATEN ISLAND"], num_shootings)

df = pd.DataFrame({"Date": np.random.choice(dates, num_shootings), "Borough": boroughs})
df.sort_values("Date", inplace=True)  # Sort by date

# Define colors
borough_colors = {
    "BROOKLYN": "blue",
    "BRONX": "red",
    "QUEENS": "green",
    "MANHATTAN": "yellow",
    "STATEN ISLAND": "purple"
}


plt.figure(figsize=(12, 6))

for borough, color in borough_colors.items():
    subset = df[df["Borough"] == borough]
    plt.scatter(subset["Date"], range(len(subset)), color=color, label=borough, alpha=0.6, s=10)


plt.xlabel("Date", fontsize=14)
plt.ylabel("Each Dot = 1 Shooting", fontsize=14)
plt.title("NYC Shootings Over Time", fontsize=16, fontweight="bold")
plt.xticks(rotation=45, fontsize=12)
plt.yticks([])

# Legend
plt.legend(title="Borough", fontsize=12)

# Show plot
plt.show()

```

This scatter plot displays shooting incidents over time, with each dot representing a single incident. Different colors for each borough were applied using color=color. The x-axis (plt.xticks(rotation=45)) was rotated for better readability of dates. The plt.yticks([  ]) was used to remove the y-axis ticks since they are not needed. A legend was added with the plt.legend function to distinguish each borough, and the font size was adjusted for clarity.

## **Plot 3: Line plot**

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample Data: Creating random shooting incidents with times
np.random.seed(42)
num_shootings = 500  # Adjust the number of incidents

# Generating random shooting times
random_seconds = np.random.randint(0, 86400, num_shootings)
times = pd.to_datetime(random_seconds, unit="s").time  # Corrected line
boroughs = np.random.choice(["BROOKLYN", "BRONX", "QUEENS", "MANHATTAN", "STATEN ISLAND"], num_shootings)

# Create DataFrame
df = pd.DataFrame({"Time": times, "Borough": boroughs})

# Convert time to numerical values (hours + fraction of an hour)
df["Time_numeric"] = df["Time"].apply(lambda t: t.hour + t.minute / 60)

# Line Chart: Incidents Over Time for Each Borough
df["Date"] = pd.date_range(start="1/1/2022", periods=len(df), freq="D")
plt.figure(figsize=(12, 6))

borough_colors = {"BROOKLYN": "blue", "BRONX": "red", "QUEENS": "green", "MANHATTAN": "orange", "STATEN ISLAND": "purple"}

for borough, color in borough_colors.items():
    borough_df = df[df["Borough"] == borough]
    borough_df.groupby(borough_df["Date"].dt.to_period("M")).size().plot(marker="o", linestyle="-", color=color, label=borough)

plt.title("Incidents Over Time by Borough")
plt.xlabel("Month")
plt.ylabel("Count")
plt.legend(title="Borough")
plt.show()
```

This line chart shows shooting incidents over time for each borough. The lines are color-coded by borough using the borough_colors dictionary, with markers (marker="o") for data points. The linestyle="-" defines solid lines, and label=borough associates each line with its corresponding borough in the legend. The title and axis labels were customized using plt.title, plt.xlabel, and plt.ylabel, ensuring clear communication.

## **Shootings Heatmap (Choropleth Map using Seaborn)**

```
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot to represent shootings per borough
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x="borough", y="shooting_count", data=nyc_geo, palette="viridis")

# Adding title and labels
plt.title("Shootings per Borough", fontsize=16, fontweight="bold")
plt.xlabel("Borough", fontsize=14)
plt.ylabel("Total Shootings", fontsize=14)

# Customizing the legend manually as Seaborn doesn't use a legend here
plt.xticks(rotation=45, ha='right')

plt.show()
```

This bar plot displays the total number of shootings per borough. The data is represented using bars, where each bar corresponds to a borough, and the height of the bar represents the number of shootings.

Properties showcased:

Palette: The palette="viridis" property is used to apply a color palette that is visually appealing and colorblind-friendly.

Fontsize and Fontweight: The fontsize=16 and fontweight="bold" properties for the title make it clear and bold.

Rotation of x-axis labels: The plt.xticks(rotation=45) rotates the borough names for better readability.

## **NYC Shootings Over Time (Scatter Plot using Seaborn)**

```
# Scatter plot to show shootings over time by borough
plt.figure(figsize=(15, 10))

# Plotting scatter plot using Seaborn
sns.scatterplot(data=df, x="Date", y=df.index, hue="Borough", palette="Set2", legend="full", s=60, alpha=0.7)

# Adding title and labels
plt.title("NYC Shootings Over Time", fontsize=16, fontweight="bold")
plt.xlabel("Date", fontsize=14)
plt.ylabel("Each Dot = 1 Shooting", fontsize=14)

# Customizing the x-axis ticks
plt.xticks(rotation=45)

# Remove y-axis ticks
plt.yticks([])

# Customize legend
plt.legend(title="Borough", fontsize=12, loc="upper left", bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

## **Incidents Over Time by Borough (Line Plot using Seaborn)**

```
# Line plot to represent incidents over time by borough
plt.figure(figsize=(19, 9))

# Plot each borough's incidents over time
sns.lineplot(data=df, x="Date", y="Time_numeric", hue="Borough", style="Borough", markers=True, dashes=False)

# Adding title and labels
plt.title("Incidents Over Time by Borough", fontsize=16, fontweight="bold")
plt.xlabel("Date", fontsize=14)
plt.ylabel("Incidents (Hours)", fontsize=14)

# Customize the legend
plt.legend(title="Borough", fontsize=12, loc="upper left", bbox_to_anchor=(1, 1))

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This line plot shows shooting incidents over time for each borough. Each line represents one borough, and the hue="Borough" ensures that each borough is plotted with a different color. style="Borough" adds different markers for each borough.

Properties showcased:

hue and style: The hue="Borough" and style="Borough" properties are used to differentiate boroughs with colors and markers. This makes the plot clear, showing distinct borough trends.

Markers and line style: The markers=True adds markers at each data point, and dashes=False makes the lines solid.

Legend customization: plt.legend positions the legend outside the plot with loc="upper left", bbox_to_anchor=(1, 1), making it easier to read without overlapping the plot.

Fontsize and rotation: The fontsize=16 and rotation=45 properties enhance readability.
